## 2. Set first column as the index.

```python
Data=Data.set_index(Data.columns[0])
Data
```

## 3. Plot all columns as time series.

```python
cols=['theta_1','theta_2','theta_3','theta_4','theta_5','theta_6']
Data[cols].plot(subplots=True,figsize=(15,10))
plt.show()
```

Plot represent our data. It is ordered by date because we set that as index.

## 4. Plot histograms of all columns, verify bin size. Plot all on a single, faceted plot (use sns.FacetGrid and melt, use sharex=False for FacetGrid) 

```python
Data2=Data.melt()
Data2
```

```python
Graph=sns.FacetGrid(Data2, col="variable", sharex=False, col_wrap=3)
Graph.map(plt.hist,"value")
```

Histograms represents our data distributions.
We can easy read from them that first three $\theta_1-\theta_3$ are more or less normally distributed with mean=0 and quite high noise.
$\theta_4$ are focused on 0 and 1 value with little noise, $\theta_5$ have also -1 values and last one $\theta_6$ is quite linearly displaced from 0 to around 12. 

## 5. Plot KDE-s (Kernel Denisty Estimators) for all columns.

```python
for i in cols:
    plt.figure()
    sns.distplot(Data[i], kde=True, hist=False)
```

KDE distribution can be interpreted as linearized histogram. Both histogram and KDE are probabiliti estimators. So for most cases it will be easier to read, but don't give us much more information than histogram plot.
